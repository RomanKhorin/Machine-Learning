```python
def encode_nominals(columns, data):
    print('Feature encoding started...\n')
    encoder = LabelEncoder()
    for name in columns:
        print('Current name: ', name)
        encoder.fit(data[name].drop_duplicates())
        labels = list(set(encoder.transform(data[name])))
        # work with binary features
        if len(labels) == 2:
            data[name] = encoder.transform(data[name])
            print("\t Feature '%s' - encoded correctly" % name)
        # work with nominal features (more than 2 values)
        else:
            dummy_features = pd.get_dummies(data[name])
            headers = dummy_features.columns.tolist()
            for header in range(len(headers)):
                if (name=='Mjob'):
                    headers[header] = 'MJ_' + headers[header]
                elif (name=='Fjob'):
                    headers[header] = 'FJ_' + headers[header]
                elif (name=='reason'):
                    headers[header] = headers[header] + '_reason'
                elif (name=='guardian'):
                    headers[header] = headers[header] + '_guardian'
            dummy_features.columns = headers
            # append new features to dataset
            data = data.join(dummy_features)
            # delete encoded feature
            data = data.drop(name, axis=1)
            print("\t Feature '%s' - encoded correctly" % name)
            
    print('\nFeature encoding finished!')
    return data
```

```python
# select nominal features
columns = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob',
           'Fjob', 'reason', 'guardian','schoolsup', 'famsup', 'paid',
           'activities', 'nursery',
       'higher', 'internet', 'romantic']

# update data with encoded features
data = encode_nominals(columns, data)
```

Введем нашу целевую переменную alc как комбинацию двух признаков Walc и Dalc с некоторыми весами.
<p>Если alc принимает значения меньше 3, то он будет принимать значение "0", иначе - "1"</p>

```python
# add target feature
data['alc'] = np.round((data.Walc*2 + data.Dalc*5)/7)

# update target feature as binary
data['alc'][data['alc'] < 3] = 0
data['alc'][data['alc'] >= 3] = 1

# update absences as binary
data['absences'][data['absences'] > 10] = 0
data['absences'][data['absences'] <= 10] = 1

data.head()
```

```python
X = scale(data[data.columns[:-1]])
y = data.alc.values
```

Проведем стандартное разбиение нашего датасета на обучающую и контрольные выборки в соотношении 80\20

```python
# split dataset on train & test
RND_SEED = 2400
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= RND_SEED)
```

Напишем метод, который будет для выбранной нами модели показывать лучшие настройки метода на кросс-валидации, лучшее значение accuracy на кросс-валидации, значение accuracy на тестовой выборке, а также будет отображать таблицу важностей признаков (по убыванию значимости) и наглядно демонстрировать важности признаков на графике.

```python
def Make_All_with_CV(model_name, param_grid):
    folds = StratifiedKFold(y_train, n_folds=5, shuffle=True, random_state=RND_SEED)
    
    if model_name=='Random Forest':
        model = GridSearchCV(RandomForestClassifier(n_estimators=25, n_jobs=-1, random_state=RND_SEED), 
                             param_grid, n_jobs=-1, cv=folds, scoring='accuracy')
    elif model_name=='Gradient Boosting':
        model = GridSearchCV(GradientBoostingClassifier(random_state=RND_SEED), param_grid,
                             n_jobs=-1, cv=folds, scoring='accuracy')
    else:
        model = GridSearchCV(AdaBoostClassifier(n_estimators=150, random_state=RND_SEED),
                             param_grid, scoring='accuracy', n_jobs=-1, cv=folds)
        
    model.fit(X_train, y_train)
    print("Best params: ", model.best_params_)
    print("Best accuracy score on CV: %f"% model.best_score_)
    print("Accuracy score on test: %f \n"% accuracy_score(y_test, model.best_estimator_.predict(X_test)))
    
    print("Feature importances for selected model:")
    importance = model.best_estimator_.feature_importances_
    plt.figure(figsize=(30,15))
    idx = np.arange(importance.size)
    width = 0.8
    p1 = plt.bar(idx, importance, width, color='c')
    plt.xticks(idx+width/2, (data.columns[:-1]))[1]
    plt.show()
    
    print('Table of feature importances for selected model: ')
    importances_table = pd.DataFrame(columns=['feature','importance'])
    importances_table.feature = data.columns[:-1]
    
    for i in range(len(importances_table.feature)):
        importances_table.importance[i] = round(importance[i]*100, 2)

    importances_table = importances_table.sort('importance', ascending=False)

    for i in range(len(importances_table.feature)):
        importances_table.importance[i] = str(round(importance[i]*100, 2))+'%'
    
    return importances_table
```

Также напишем функцию, которая будет определять значения accuracy для 5 лучших настроек модели

```python
def Find_best_accuracies(model_name, param_grid):
    best_models, best_accuracies = [], []
    
    for fold in [2, 3, 4, 5, 6]:
        folds = StratifiedKFold(y_train, n_folds=fold, shuffle=True, random_state=RND_SEED)

        if model_name=='Random Forest':
            model = GridSearchCV(RandomForestClassifier(n_estimators=25, n_jobs=-1, random_state=RND_SEED),
                                 param_grid, n_jobs=-1, cv=folds, scoring='accuracy')
        elif model_name=='Gradient Boosting':
            model = GridSearchCV(GradientBoostingClassifier(random_state=RND_SEED), param_grid,
                                 n_jobs=-1, cv=folds, scoring='accuracy')
        else:
            model = GridSearchCV(AdaBoostClassifier(n_estimators=150, random_state=RND_SEED),
                                 param_grid, scoring='accuracy', n_jobs=-1, cv=folds)
        
        model.fit(X_train, y_train)
        best_models.append(model.best_estimator_)
    
    for model in best_models:
        best_accuracies.append(accuracy_score(y_test, model.predict(X_test)))
        
    return best_accuracies, best_models
```

## Random Forest

```python
rf_param_grid = {'criterion' : ['gini', 'entropy'],
             'max_features': ['auto', 'sqrt', 'log2'],
             'max_depth' : np.arange(1, 21, 1),
             'class_weight' : [None, 'balanced']
             }

Make_All_with_CV('Random Forest', rf_param_grid)
```

```python
%%time
best_accuracies_rf, best_rfs = Find_best_accuracies('Random Forest', rf_param_grid)
```

#### Значение accuracy для 5 лучших RF

## Gradient Boosting

```python
gb_param_grid = {'loss':['deviance', 'exponential'],
              'max_features':['auto', 'log2', None],
              'max_depth': np.arange(1,17,1)
             }

Make_All_with_CV('Gradient Boosting', gb_param_grid)
```

Причина, по которой важности признаков GB отличаются от RF может быть в том, что в RF каждое дерево обучается на случайной подвыборке данных и важность признаков становится средневзвешенной, в то время
как в GB важность признаков исправляется после добавления каждого нового дерева в композицию

```python
%%time
best_accuracies_gb, best_gbs = Find_best_accuracies('Gradient Boosting', gb_param_grid)
```
